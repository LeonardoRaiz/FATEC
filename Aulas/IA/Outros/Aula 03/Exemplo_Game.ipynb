```python
import random
import time

class BossNPC:
    def __init__(self):
        self.state = "PATROL" 
        self.player_distance = 50 
        self.hp = 100

    def update(self):
        
        self.player_distance = random.randint(1, 60)
        print(f"\n[SENSOR] Dist√¢ncia do Jogador: {self.player_distance}m | Estado Atual: {self.state}")

        # L√≥gica da FSM (C√©rebro do Agente)
        if self.state == "PATROL":
            if self.player_distance < 20:
                print("‚ö†Ô∏è Jogador detectado! Mudando para PERSEGUI√á√ÉO!")
                self.state = "CHASE"
            else:
                print("üéµ Assobiando... patrulhando a √°rea.")

        elif self.state == "CHASE":
            if self.player_distance < 5:
                print("‚öîÔ∏è Perto o suficiente! Mudando para ATAQUE!")
                self.state = "ATTACK"
            elif self.player_distance > 30:
                print("‚ùì Perdi ele de vista... voltando para PATRULHA.")
                self.state = "PATROL"
            else:
                print("üèÉ CORRA COVARDE! Perseguindo...")

        elif self.state == "ATTACK":
            if self.player_distance > 5:
                self.state = "CHASE"
            else:
                print("üî• SMASH! Dano massivo no jogador!")


boss = BossNPC()
for frame in range(10): 
    boss.update()
    time.sleep(1) 
```
