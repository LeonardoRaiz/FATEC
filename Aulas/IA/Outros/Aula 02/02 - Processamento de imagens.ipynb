```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Defina o nome do arquivo da imagem
nome_arquivo = "ressonancia.jpg"

# 2. Carregue a imagem do disco
# A imagem é carregada em cores por padrão (formato BGR)
imagem_original = cv2.imread(nome_arquivo)

# 3. Verifique se a imagem foi carregada com sucesso
if imagem_original is None:
    print(f"Erro: não foi possível encontrar o arquivo '{nome_arquivo}'.")
    print("Verifique se o arquivo está na mesma pasta que o seu script.")
else:
    # 4. Converta a imagem para tons de cinza
    imagem_cinza = cv2.cvtColor(imagem_original, cv2.COLOR_BGR2GRAY)

    # 5. Mostre a imagem e a matriz de pixels
    plt.figure(figsize=(10, 4))

    # Primeiro gráfico: a imagem
    plt.subplot(1, 2, 1)
    plt.imshow(imagem_cinza, cmap="gray")
    plt.title("Imagem em Tons de Cinza")
    plt.axis("off")

    # Segundo gráfico: a matriz de pixels sobreposta
    plt.subplot(1, 2, 2)
    plt.imshow(imagem_cinza, cmap="gray")
    plt.title("Matriz de Pixels")

    # Exibe os valores dos pixels com um espaçamento de 30 para não poluir
    for i in range(0, imagem_cinza.shape[0], 30):
        for j in range(0, imagem_cinza.shape[1], 30):
            plt.text(j, i, str(imagem_cinza[i, j]), color="red", fontsize=6, ha='center', va='center')

    plt.show()
```
