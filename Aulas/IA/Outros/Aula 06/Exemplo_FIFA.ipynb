```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Gerando Dados SintÃ©ticos de Jogadores (Dataset "FIFA 2026")
# Atributos: [Velocidade, Chute, Drible, Defesa]
np.random.seed(42)
n_players = 500

velocidade = np.random.randint(50, 99, n_players)
chute = np.random.randint(40, 99, n_players)
drible = np.random.randint(50, 99, n_players)

# A "FÃ³rmula Secreta" do valor de mercado (com um pouco de aleatoriedade)
# Valor = (Velocidade * 1000) + (Chute * 2000) + (Drible * 1500) - 50000
valor_real = (velocidade * 1000) + (chute * 2000) + (drible * 1500) - 50000 + np.random.randint(-5000, 5000, n_players)

df = pd.DataFrame({'Pace': velocidade, 'Shooting': chute, 'Dribbling': drible, 'Value_Euros': valor_real})

# 2. Treinando a IA (O Agente Esportivo)
X = df[['Pace', 'Shooting', 'Dribbling']]
y = df['Value_Euros']

model = LinearRegression()
model.fit(X, y)

print("ðŸ¤– Modelo Treinado! Aprendi como o mercado funciona.")

# 3. Fazendo uma PrediÃ§Ã£o (Scouting)
# Novo jogador promessa: Muito rÃ¡pido (95), bom drible (88), chute mÃ©dio (70)
novo_jogador = np.array([[95, 70, 88]])
valor_previsto = model.predict(novo_jogador)

print(f"\nâš½ Stats do Jogador: Pace 95, Shooting 70, Dribbling 88")
print(f"ðŸ’° Valor Justo Estimado pela IA: â‚¬{valor_previsto[0]:,.2f}")
print(f"ðŸ§  Pesos (ImportÃ¢ncia): Pace={model.coef_[0]:.1f}, Shooting={model.coef_[1]:.1f}, Dribbling={model.coef_[2]:.1f}")
```
